#Vertical Bar Plot (Biểu đồ thanh dọc)

```python
import numpy as np
import matplotlib.pyplot as plt
marks=[72,40,33,90,100]
bars=("Roll 1",'Roll 2',"Roll 3",'Roll 4',"Roll 5")
y=np.arange(len(bars))
plt.bar(y,marks,color="r");
```

#Horizontal Bar Plot ( Biểu đồ thanh ngang)

```python
import numpy as np
import matplotlib.pyplot as plt
marks=[72,40,33,90,100]
bars=["Roll 1","Roll 2","Roll 3","Roll 4","Roll 5"]
y=np.arange(len(bars))
plt.barh(y,marks,color='b')
plt.yticks(y,bars);
```

#Trực quan hóa doanh thu (Triệu đồng) của các Cửa hàng trong tháng

```python
x=("Cửa hàng 1","Cửa hàng 2","Cửa hàng 3","Cửa hàng 4",'Cửa hàng 5')
y=([65,40,38,85,100])
plt.bar(x,y,color="y")
plt.title("Doanh số bán trong tháng")
plt.xlabel("Số lượng cửa hàng")
plt.ylabel("Doanh thu đạt được(Triệu đồng)")
plt.show()
```

#Ví dụ 3: Bar thể hiện Nhóm tuổi của Male và Female

```python
import matplotlib.pyplot as plt
import numpy as np
Male=[30,34,37,35,27]
Female=[25,32,34,40,25]
group=["Nhóm 1","Nhóm 2","Nhóm 3","Nhóm 4","Nhóm 5"]
#Lấy cột từ group
x=np.arange(len(group))
#Độ rộng của bar
width=0.5
fig,ax=plt.subplots()
rects1=ax.bar(x-width/2 ,Male,width,color="g",label="Male")
rects2=ax.bar(x+width/2,Female,width,color="b",label="Female")
#Gán Label và Title
ax.legend()
ax.set_ylabel('Độ tuổi')
ax.set_title("Nhóm tuổi và giới tính")
ax.set_xticks(x)
ax.set_xticklabels(group)
```
