```python
import matplotlib.pyplot as plt

# dataset-1
x1 = [89, 43, 36, 36, 95, 10, 87,
    66, 34, 38, 20, 26, 34, 90, 33]
y1 = [21, 46, 3, 26, 34, 90, 33, 35,
      53, 72, 58, 10, 67, 95, 120]
# dataset2
x2 = [26, 29, 48, 20, 56, 2, 48, 20, 56,
    36, 66, 72, 20, 56, 2, 47, 40]
y2 = [26, 34, 90, 33, 38, 42, 20, 8,
      20, 56, 2, 47, 15, 47, 64, 6, 5]
#Tạo các nốt với các thuộc tính sau
plt.scatter(x1,y1,c="pink",
            linewidths=2,
            marker="s",
            edgecolor="red",
            s=100)
plt.scatter(x2,y2,c="yellow",
            linewidths=2,
            marker="^",
            edgecolor="green",
            s=150)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu ngẫu nhiên
n1, n2, n3, n4,y = np.random.rand(20), np.random.rand(20), np.random.rand(20), np.random.rand(20),np.random.rand(20)

# Vẽ biểu đồ scatter với 4 màu và kiểu khác nhau
plt.scatter(y, n1, color='blue', marker='^',label="blue")
plt.scatter(y, n2, color='green', marker='*',label="green")
plt.scatter(y, n3, color='red', marker='v',label="red")
plt.scatter(y, n4, color='orange', marker='o',label="orange")

# Thêm tiêu đề và chú thích
plt.title('Biểu đồ Scatter với 4 màu và kiểu khác nhau')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='lower right',ncol=2)
# Hiển thị biểu đồ
plt.show()
```

```python
import numpy as np
import pandas as pd
#Hàm khởi tạo các bộ sinh số ngẫu nhiên (random generator)
#Biến số trong seed thường là một số nguyên không âm 32 bit
np.random.seed(12456)
#Gán số lượng N (Bong bóng)
N=75
x=np.random.rand(N)
y=np.random.rand(N)
#Gán màu sắc random
colors=np.random.rand(N)
#Gán bán kính random cho các điểm từ 0 đến 40 (Độ lớn bình phương)
area=(40*np.random.rand(N))**2
#In theo x và y (diện tích và màu sắc)
plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
```
