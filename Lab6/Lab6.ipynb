##Bài 1: Tạo Danh sách các mảng là đầu vào bắt buộc để tạo boxplot.

```python
import numpy as np
import matplotlib.pyplot as plt
#Tao mang ngau nhien
np.random.seed(10)
collectn_1=np.random.normal(100,10,200)
collectn_2=np.random.normal(80,30,200)
collectn_3=np.random.normal(90,20,200)
collectn_4=np.random.normal(70,25,200)
fig=plt.figure()
#tao 1 ax
ax=fig.add_axes([0,0,1,1])
#Tao boxplot
bp=ax.boxplot([collectn_1,collectn_2,collectn_3,collectn_4])
plt.show()
```

##Bài 2: Biểu đồ Box plot cơ bản:

```python
import numpy as np
import matplotlib.pyplot as plt
#Trang thai ngau nhien co kha nang tai tao
np.random.seed(163377555)
#Tao 1 du lieu gia
spread=np.random.rand(50)*100
center=np.ones(25)*50
flier_high=np.random.rand(10)*100+100
flier_low=np.random.rand(10)*-100
data=np.concatenate((spread,center,flier_high,flier_low))
fig1,ax1=plt.subplots()
ax1.set_title("Basic Plot")
ax1.boxplot(data)
plt.show()
```

##Bài 3: Sử dụng hàm boxplot() của seaborn. Nhìn vào biểu đồ, sẽ dễ dàng kết luận rằng nhóm ‘C’ có giá trị cao hơn các nhóm khác. Tuy nhiên, ta không thể thấy sự phân bố cơ bản của các dấu chấm trong mỗi nhóm, cũng như số lượng quan sát cho mỗi nhóm.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
#Dataset
a=pd.DataFrame({"group":np.repeat("A",500),"value":np.random.normal(10,5,500)})
b=pd.DataFrame({"group":np.repeat("B",500),"value":np.random.normal(13,1.2,500)})
c=pd.DataFrame({"group":np.repeat("B",500),"value":np.random.normal(18,1.2,500)})
d=pd.DataFrame({"group":np.repeat("C",20),"value":np.random.normal(25,4,20)})
e=pd.DataFrame({"group":np.repeat("D",100),"value":np.random.uniform(12,size=100)})
df=a.append(b).append(c).append(d).append(e);
color = ("blue","red","green","pink")
#Bieu dien boxplot
sns.boxplot(x="group",y="value",data=df,palette=color);
```

##Bai 4:Phân phối điểm kiểm tra cho tất cả các môn học trong một biểu đồ với dữ liệu ngẫu nhiên
